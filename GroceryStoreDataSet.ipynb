```python
!pip install pandas mlxtend

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('GroceryStoreDataSet.csv')
df.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

data = []

for row in df.values:
    # ambil hanya yang tidak kosong (NaN)
    transaksi = [str(item) for item in row if pd.notna(item)]
    data.append(transaksi)

te = TransactionEncoder()
te_data = te.fit(data).transform(data)

df = pd.DataFrame(te_data, columns=te.columns_)
df.head()
```

```python
frequent = apriori(df, min_support=0.02, use_colnames=True)
frequent.sort_values(by='support', ascending=False)
```

```python
rules = association_rules(frequent, metric="confidence", min_threshold=0.3)
rules.sort_values(by='confidence', ascending=False)
```
